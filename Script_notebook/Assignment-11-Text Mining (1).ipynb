## Named Entity Recognition (NER)

```python
nlp=spacy.load('en_core_web_sm')

one_block=clean_tweets
doc_block=nlp(one_block)
spacy.displacy.render(doc_block,style='ent',jupyter=True)
```

```python
for token in doc_block[100:200]:
    print(token,token.pos_) 
```

```python
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[100:200])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10]
```

```python
wd_df[0:10].plot.bar(x='word',figsize=(12,8),title='Top 10 nouns and verbs');
```

## Importing Affin

```python
afinn = pd.read_csv("C:/Users/BASHAR/OneDrive/Documents/Latest Data Science Material/Day-34/Afinn.csv", encoding="latin-1")
afinn.head()
```

```python
from itertools import islice

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))
```

```python
affin_score = afinn.set_index("word")["value"].to_dict()
take(20, affin_score.items())
```

```python
nlp = spacy.load("en_core_web_sm")
sentiment_lexicon = affin_score

def calculate_sentiment(text: str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
sent_df["sentiment_score"] = sent_df["sentence"].apply(calculate_sentiment)
```

```python
sent_df["Word_count"] = sent_df["sentence"].str.split().apply(len)
sent_df.head()
```

```python
sent_df.sort_values(by="sentiment_score")
```
